```python
import pandas as pd
import requests

url = 'https://api.census.gov/data/2019/acs/acs5?get=B01001_001E,B17001_002E,B17001_003E,GEO_ID,NAME&for=tract:*&in=state:36&in=county:*'
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data[1:], columns=data[0])
```

```python
df
```

```python
df.rename(columns={'B01001_001E': 'Total Population', 'B17001_002E': 'Below Poverty Level', 'B17001_003E': 'Above Poverty Level', 'NAME': 'Name', 'GEO_ID': 'geoid'}, inplace=True)
```

```python
df
```

```python
df[['Census Tract', 'County Name', 'State']] = df['Name'].str.split(',', expand=True)
df
```

```python
df.drop(columns=['Name', 'state', 'County Name', 'State', 'Census Tract'], inplace=True)
df
```

```python
counties = ['047', '005', '061', '085', '081']
county = df['county'].isin(counties)
df = df[county]
df
```

```python
df['county'] = df['county'].replace({'047':'Kings', '005': 'Bronx', '061': 'New York', '085': 'Richmond', '081': 'Queens'})
```

```python
df
```

```python
df.to_csv('Poverty.csv', index=False)
```

```python
import numpy as np
# Define a list of column names to calculate weighted average for
cols_to_average = ['Below Poverty Level', 
                   'Above Poverty Level', 'Less than 9th grade', '9th to 12th grade, no diploma', 
                   'High school diploma', 'Some college', 'Associate degree', 'Bachelor degree', 
                   'Graduate degree', 'High school graduate, or higher', 'Bachelor degree, or higher']

# Calculate the weighted average for each column based on the 'Total Population' column
weighted_average = m[cols_to_average].apply(lambda x: np.average(x, weights=m['Total Population']))

# Add the resulting weighted averages as new columns in the DataFrame
df_weighted = m.assign(**weighted_average)

# Print the resulting DataFrame with the weighted averages as new columns
df_weighted
```

```python
df_weighted.rename(columns={'Below Poverty Level': 'Below Poverty Level WA', 
'Above Poverty Level': 'Above Poverty Level WA', 
'Less than 9th grade': 'Less than 9th grade WA',
'9th to 12th grade, no diploma': '9th to 12th grade, no diploma WA', 
'High school diploma': 'High school diploma WA',
'Some college' : 'Some college WA',
'Associate degree': 'Associate degree WA', 
'Bachelor degree': 'Bachelor degree WA', 
'Graduate degree': 'Graduate degree WA',
'High school graduate, or higher': 'High school graduate, or higher WA', 
'Bachelor degree, or higher': 'Bachelor degree, or higher WA'}, inplace=True)
```
